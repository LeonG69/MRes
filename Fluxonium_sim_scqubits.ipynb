```python
import numpy as np
import scqubits as scq
```

```python
fluxonium = scq.Fluxonium(EJ = 7,
                               EC = 1,
                               EL = 1,
                               flux = 0.5,
                               cutoff = 110)
```

```python
fluxonium.eigenvals()
```

```python
flux_list = np.linspace(-0.5, 0.5, 151)
fluxonium.plot_evals_vs_paramvals('flux', flux_list, evals_count=8);
```

```python
fluxonium.flux = 0.5
fluxonium.plot_wavefunction(esys=None, which=range(5), mode='real');
```

```python
fluxonium = scq.Fluxonium(
    EJ=7,
    EC=1,
    EL=1,
    cutoff = 100,
    flux = 0.5
)
fig, axes=fluxonium.plot_wavefunction(esys=None, which=[0,1,2,3], mode='real');
```

```python
phimat = fluxonium.matrixelement_table('phi_operator', evals_count=3)
print(phimat)

```

```python
nmat = fluxonium.matrixelement_table('n_operator', evals_count=3)
print(nmat)
```

```python
fluxonium.plot_matelem_vs_paramvals('n_operator', 'flux', flux_list, select_elems=3);
```
